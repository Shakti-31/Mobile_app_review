```python
figure = plt.figure(figsize=(20, 5))
sns.barplot
ax = sns.countplot(x='app_name' , data =data)
for bars in ax.containers:
  ax.bar_label(bars)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='app_category' , data =data, palette='rocket')
for bars in ax.containers:
  ax.bar_label(bars)
  plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
user_country = data['user_gender'].value_counts(normalize=True)
print(user_country)

plt.figure(figsize=(5,4))
plt.pie(user_country, labels=user_country.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
user_country = data['verified_purchase'].value_counts(normalize=True)
print(user_country)

plt.figure(figsize=(5,4))
plt.pie(user_country, labels=user_country.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```
